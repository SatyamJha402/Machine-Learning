```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Basic Graph

```python
# plt.plot([1,2,3,4],[1,4,9,16])

x=[1,2,3,4]
y=[1,8,27,64]
plt.plot(x,y, label="My Graph", color="y", linewidth=5, marker="*", markersize="15", markeredgecolor="blue", linestyle="--")
# plt.plot(x,y, "y*--", label="2x")
# format="[color][marker][linestyle]"
plt.title("Graph", fontdict={"fontname": "Comic Sans MS", "fontsize":"20"})



plt.xlabel("X")
plt.ylabel("Y")

plt.legend()

# plt.xticks([1,2,3,10])

plt.show()
```

```python
plt.figure(figsize=(2,1.5), dpi=100)
x2=np.arange(0,10,1)
plt.plot(x2[:6], x2[:6]**2, "r>-")
plt.plot(x2[5:], x2[5:]**2, "m<--")
plt.savefig("myGraph", dpi=300)
plt.show()
```
